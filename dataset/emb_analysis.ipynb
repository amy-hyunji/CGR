```python
def get_top(emb, _emb, top=10):
    score = torch.inner(_emb, emb)
    topk = torch.topk(score, 10)
    return topk.values, np.array(topk.indices)

def remove_punc(text):
    exclude = set(string.punctuation)
    exclude.add("▁")
    return "".join(ch for ch in text if ch not in exclude)
        
head = 100
_head = 50
models = ["Enc-Dec", "Enc-mean", "Enc-first"]
for ite, (id, emb) in enumerate(clusterId_emb.items()):
    #print(ite)
    if id == 0 or id == 1: continue
    if ite % 10 == 0: print(ite)
    if(ite < 2+_head): continue
    _head += 1
    
    if _head > head: break
    text = clusterId_tokText[id]
    print('='*50)
    print(f'id: {id}\ntext: {text}')
    print(' ')
    text = remove_punc(text)
    
    print('## [Enc-Dec]', '-'*50)
    emb = clusterId_emb[id]
    values, indices = get_top(torch.tensor(emb), d_emb)
    for val, ind in zip(values, indices):
        _text = clusterId_tokText[ind]
        _group_id = clusterId_tokGroupId[ind]
        _tokIdList = tokGroupId_tokIdList[_group_id]
        _isContain = False
        for _tokId in _tokIdList:
            _title = tokId_corpus[_tokId]
            if text in _title:
                _isContain = True
                break
        print(f'id: {ind}\ttext: {_text}\tisContain: {_isContain}\tscore: {round(float(val), 2)}')
    print(' ')
    
    text = m_clusterId_tokText[id]
    print("text:", text)
    text = remove_punc(text)
    print('## [From -Mean]', '-'*50)
    emb = m_clusterId_emb[id]
    values, indices = get_top(torch.tensor(emb), m_emb)
    for val, ind in zip(values, indices):
        _text = clusterId_tokText[ind]
        _group_id = clusterId_tokGroupId[ind]
        _tokIdList = tokGroupId_tokIdList[_group_id]
        _isContain = False
        for _tokId in _tokIdList:
            _title = tokId_corpus[_tokId]
            if text in _title:
                _isContain = True
                break
        print(f'id: {ind}\ttext: {_text}\tisContain: {_isContain}\tscore: {round(float(val), 2)}')
    print(' ')
    
    text = f_clusterId_tokText[id]
    print("text:", text)
    text = remove_punc(text)
    print('## [From -First]', '-'*50)
    emb = f_clusterId_emb[id]
    values, indices = get_top(torch.tensor(emb), f_emb)
    for val, ind in zip(values, indices):
        _text = clusterId_tokText[ind]
        _group_id = clusterId_tokGroupId[ind]
        _tokIdList = tokGroupId_tokIdList[_group_id]
        _isContain = False
        for _tokId in _tokIdList:
            _title = tokId_corpus[_tokId]
            if text in _title:
                _isContain = True
                break
        print(f'id: {ind}\ttext: {_text}\tisContain: {_isContain}\tscore: {round(float(val), 2)}')
    print(' ')
    
    print('='*50)
    print('\n\n')
```
